```python
from sklearn import svm
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
iris = pd.read_csv("/content/IRIS.csv")
```

```python
iris.head()
```

```python
iris.describe()
```

```python
train, test = train_test_split(iris,test_size =0.25 )
```

```python
train
```

```python
test
```

```python
train_x = train[['sepal_length', 'sepal_width', 'petal_width', 'petal_width'] ]
```

```python
train_y = train.species
```

```python
train_x
```

```python
train_y
```

```python
test_x = test[['sepal_length', 'sepal_width', 'petal_width', 'petal_width']]
test_y = test.species
```

```python
#SVM from sklean
model = svm.SVC()
```

```python
model.fit(train_x , train_y)
pred = model.predict(test_x)
```

```python
#predict values will be printed here
```

```python
pred
```

```python
metrics.accuracy_score(pred, test_y)
```

```python
#DecisionTree
model = DecisionTreeClassifier()
```

```python
model.fit(train_x , train_y)
```

```python
pred = model.predict(test_x)
```

```python
metrics.accuracy_score(pred, test_y)
```
